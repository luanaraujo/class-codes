Escolha duas imagens, de preferência do mesmo tamanho, e façam as seguintes operações:
 - SOMA
 - SUBTRAÇÃO
 - UNIÃO

 

```python
# TODO: carregar as libs para o exercício# TODO: plotar as imagens lado a lado: original,
```

```python
import numpy as np
from PIL import Image
import requests
from io import BytesIO
from matplotlib import image
from matplotlib import pyplot as plt
```

```python
# TODO: carregar as imagens que serão utilizadas
```

```python
response = requests.get('https://wallpapercave.com/wp/wp4712331.jpg')
image1 = Image.open(BytesIO(response.content)).convert()
image1 = np.asarray(image1)

plt.figure(figsize=(16,9))
plt.imshow(image1)
plt.axis('off')
plt.show()

response = requests.get('https://eskipaper.com/images/passenger-airplanes-9.jpg')
image2 = Image.open(BytesIO(response.content)).convert()
image2 = np.asarray(image2)

plt.figure(figsize=(16,9))
plt.imshow(image2)
plt.axis('off')
plt.show()
```

```python
# TODO: somar imagem1 com imagem2, e plotar o resultado
```

```python
soma = image1 + image2
plt.axis('off')
plt.imshow(soma)
```

```python
# TODO: somar imagem2 com imagem1, e plotar o resultado
```

```python
soma2 = image2 + image1
plt.axis('off')
plt.imshow(soma2)
```

```python
# TODO: plotar as imagens lado a lado: original, (img1 + img2) e (img2 + img1) 
```

```python
fig, (a, b, c, d) = plt.subplots(1, 4, figsize=(16,9))

a.set_title('Imagem 1 Original')
a.axis('off')
a.imshow(image1)

b.set_title('Imagem 2 Original')
b.axis('off')
b.imshow(image2)

c.set_title('Imagem 1 + Imagem 2')
c.axis('off')
c.imshow(soma)

d.set_title('Imagem 2 + Imagem 1')
d.axis('off')
d.imshow(soma2)

plt.show()
```

```python
# TODO: subtrair imagem1 com imagem2, e plotar o resultado
```

```python
sub = image1 - image2
plt.axis('off')
plt.imshow(sub)
```

```python
# TODO: subtrair imagem2 com imagem1, e plotar o resultado
```

```python
sub2 = image2 - image1
plt.axis('off')
plt.imshow(sub2)
```

```python
# TODO: plotar as imagens lado a lado: original, (img1 + img2) e (img2 + img1) 
```

```python
fig, (a, b, c, d) = plt.subplots(1, 4, figsize=(16,9))

a.set_title('Imagem 1 Original')
a.axis('off')
a.imshow(image1)

b.set_title('Imagem 2 Original')
b.axis('off')
b.imshow(image2)

c.set_title('Imagem 1 + Imagem 2')
c.axis('off')
c.imshow(soma)

d.set_title('Imagem 2 + Imagem 1')
d.axis('off')
d.imshow(soma2)

plt.show()
```

```python
# TODO: aplicar a operação de união imagem1 com imagem2, e plotar o resultado
```

```python
uniao = image2 * image1
plt.axis('off')
plt.imshow(uniao)
```

```python
# TODO: plotar as imagens lado a lado: original e união
```

```python
fig, (a, b, c) = plt.subplots(1, 3, figsize=(16,9))

a.set_title('Imagem 1 Original')
a.axis('off')
a.imshow(image1)

b.set_title('Imagem 2 Original')
b.axis('off')
b.imshow(image2)

c.set_title('União')
c.axis('off')
c.imshow(uniao)
```
