```python
import numpy as np
from PIL import Image
import requests
from io import BytesIO
from matplotlib import pyplot as plt
```

CARREGAR A IMAGEM

```python
response = requests.get('https://images3.alphacoders.com/266/thumb-1920-266807.jpg')
img1 = Image.open(BytesIO(response.content)).convert()
img1 = np.asarray(img1)

plt.figure(figsize=(16,9))
plt.imshow(img1)
plt.axis('off')
plt.show()
```
